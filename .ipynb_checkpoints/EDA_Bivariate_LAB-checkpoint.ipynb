```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
df = pd.read_csv("Data assets/amz_uk_price_prediction_dataset.csv") 
```

# Part 1

```python
cat_isbest = pd.crosstab(df['category'], df['isBestSeller'])
cat_isbest['Proportion'] = round((cat_isbest[True] / (cat_isbest[True] + cat_isbest[False])) * 100, 2)
cat_isbest_sorted = cat_isbest.sort_values(by='Proportion', ascending=False)
cat_isbest_sorted
```

```python
cat_isbest = pd.crosstab(df['category'], df['isBestSeller'])
```

```python
from scipy.stats import chi2_contingency

chi2_statistic, chi2_p_value, _, _ = chi2_contingency(cat_isbest)

chi2_statistic, chi2_p_value
```

```python
from scipy.stats.contingency import association

association(cat_isbest, method="cramer")
```

# Part 2

```python
Q1 = df['price'].quantile(0.25)  # First quartile (25th percentile)
Q3 = df['price'].quantile(0.75)  # Third quartile (75th percentile)
IQR = Q3 - Q1  # Interquartile range

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df_no_outliers = df[(df['price'] >= lower_bound) & (df['price'] <= upper_bound)]

df_no_outliers
```

```python
category_counts = df_no_outliers['category'].value_counts().sort_values(ascending=False).head(20).index

df_filtered = df_no_outliers[df_no_outliers['category'].isin(category_counts)]

plt.figure(figsize=(20, 8)) 
sns.violinplot(data=df_filtered, x='category', y='price', palette="coolwarm")

plt.xticks(rotation=45, ha='right') 
plt.show()
```

```python
top_10_categories = df_no_outliers['category'].value_counts(ascending=False).head(10).index

df_filtered = df_no_outliers[df_no_outliers['category'].isin(top_10_categories)]

avg_price_top_10 = df_filtered.groupby('category')['price'].mean().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(x=avg_price_top_10.index, y=avg_price_top_10.values, palette="coolwarm")

plt.xticks(rotation=45, ha='right')
```

```python
df_stars = df_filtered[df_filtered['stars'] != 0]
plt.figure(figsize=(20,8))
sns.boxplot(data=df_stars, x='category', y='stars', palette="coolwarm")
plt.xticks(rotation=90)
plt.show()
```
